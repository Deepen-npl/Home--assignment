```python
from itertools import permutations
from itertools import combinations
```

```python
str_1='ATGC'#string is the list of characters
print(str_1) #also can be used str_1= ['A','T','G','C']
```

```python
perm = [] #open empty list
for r in range(1,4):#range of start 1, to 4-1=3
    c = permutations(str_1,r)
    for i in c:
        el = ''.join(list(i)) # join all string of perm
        perm.append(el)
print(perm)
```

```python
for each in perm:
    print(f'the number of {each} in DNA is {DNA.count(each)}')
```
